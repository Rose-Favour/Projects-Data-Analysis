```python
#identifying loyal customers
#using the RFM method-recency,frequecny and monetary
#We filter customers for chicago

shop_chicago = online_shop3[online_shop3.Location=='Chicago']
shop_chicago.head(5)
```

```python
shop_chicago.describe()
```

```python
shop_chicago.loc[:,'Transaction_Date']=pd.to_datetime(shop_chicago.loc[:,'Transaction_Date'])

```

```python
recency = shop_chicago.groupby(by='CustomerID', 
                        as_index=False)['Transaction_Date'].max()
recency.columns = ['CustomerID', 'PurchaseDate']
recent_date = recency['PurchaseDate'].max()
recency['Recency'] = recency['PurchaseDate'].apply(
    lambda x: (recent_date - x).days)
recency.head(5)
```

```python
frequency = shop_chicago.drop_duplicates().groupby(by=
    ['CustomerID'],as_index=False)['Transaction_ID'].count()
frequency.columns = ['CustomerID', 'Frequency']
frequency.head()
```

```python
monetary = shop_chicago.groupby(by='CustomerID', as_index=False)['Sales'].sum()
monetary.columns = ['CustomerID', 'Monetary']
monetary.head()
```

```python
#Merging the three columns
rfm = recency.merge(frequency, on='CustomerID')
rfm = rfm.merge(monetary, on='CustomerID').drop(
	columns='PurchaseDate')
rfm.head()
```

```python
#Ranking
rfm['R_rank'] = rfm['Recency'].rank(ascending=False)
rfm['F_rank'] = rfm['Frequency'].rank(ascending=True)
rfm['M_rank'] = rfm['Monetary'].rank(ascending=True)
#Normalization -converting to %
rfm['R_rank_norm'] = (rfm['R_rank']/rfm['R_rank'].max())*100
rfm['F_rank_norm'] = (rfm['F_rank']/rfm['F_rank'].max())*100
rfm['M_rank_norm'] = (rfm['F_rank']/rfm['M_rank'].max())*100
 
rfm.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
rfm
```

```python
rfm['RFM_Score'] = 0.15*rfm['R_rank_norm']+0.28 *  \
 rfm['F_rank_norm']+0.57*rfm['M_rank_norm'] 
	
rfm['RFM_Score'] *= 0.05
rfm = rfm.round(2)
rfm[['CustomerID', 'RFM_Score']].tail(5)
```

```python
import numpy as np
```

```python
rfm["Customer_segment"] = np.where(rfm['RFM_Score'] >
                                      4.5, "Top Customers", 
                                      (np.where(
                                        rfm['RFM_Score'] > 4,
                                        "High value Customer",
                                        (np.where(
    rfm['RFM_Score'] > 3,
                             "Medium Value Customer",
                             np.where(rfm['RFM_Score'] > 1, 
                            'Low Value Customers', 'Lost Customers'))))))
rfm[['CustomerID', 'RFM_Score', 'Customer_segment']].head(20)
```

```python
plt.pie(rfm.Customer_segment.value_counts(),
        labels=rfm.Customer_segment.value_counts().index,
        autopct='%.0f%%')
plt.show()
```
